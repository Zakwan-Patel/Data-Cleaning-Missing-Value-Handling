```python
import pandas as pd
import numpy as np

# Load dataset
house_df = pd.read_csv("House_Prices.csv")  

# Check missing values
missing_before = house_df.isnull().sum().sort_values(ascending=False)
missing_before[missing_before > 0]
```

```python
import matplotlib.pyplot as plt

missing_data = missing_before[missing_before > 0]
if len(missing_data) > 0:
    missing_data.plot(
        kind="bar",
        figsize=(12,5),
        title="Missing Values per Column (Before Cleaning)"
    )
    plt.ylabel("Count")
    plt.show()
else:
    print("No missing values found in the dataset")
```

```python
num_cols = house_df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = house_df.select_dtypes(include=["object"]).columns
```

```python
threshold = 0.6 * len(house_df)  # 60% missing
cols_to_drop = house_df.columns[house_df.isnull().sum() > threshold]

house_df.drop(columns=cols_to_drop, inplace=True)
```

```python
for col in num_cols:
    if col in house_df.columns:
        house_df[col].fillna(house_df[col].median(), inplace=True)
```

```python
for col in cat_cols:
    if col in house_df.columns:
        house_df[col].fillna(house_df[col].mode()[0], inplace=True)
```

```python
print("Before Cleaning Shape:", missing_before.shape)
print("After Cleaning Shape:", house_df.shape)
```

```python
house_df.to_csv("House_Prices_Cleaned.csv", index=False)
```
