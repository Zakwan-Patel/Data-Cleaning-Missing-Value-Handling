```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

medical_df = pd.read_csv("Medical_Appointment.csv")

medical_df.isnull().sum()
```

```python
medical_df.isnull().sum()[medical_df.isnull().sum()].plot(
    kind="bar",
    figsize=(8,4),
    title="Missing Values (Medical Dataset)"
)
plt.show()
```

```python
num_cols_med = medical_df.select_dtypes(include=["int64", "float64"]).columns
cat_cols_med = medical_df.select_dtypes(include=["object"]).columns
```

```python
for col in num_cols_med:
    medical_df[col].fillna(medical_df[col].median(), inplace=True)
```

```python
for col in cat_cols_med:
    medical_df[col].fillna(medical_df[col].mode()[0], inplace=True)
```

```python
medical_df.isnull().sum().any()
```

```python
medical_df.to_csv("Medical_Appointments_Cleaned.csv", index=False)
```
